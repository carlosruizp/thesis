```python
# imports
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression

from mlxtend.plotting import plot_decision_regions
```

```python
n = 100
xy_min = [-2, -8]
xy_max = [2, 8]
xlinsp = np.linspace(-2, 2, 100)
data = np.random.uniform(low=xy_min, high=xy_max, size=(n,2))
```

```python
f = (lambda x: x**3)
f_sign = (lambda x, y: np.sign(y - x**3))
```

```python
labels = f_sign(data[:, 0], data[:, 1]).astype(int)
# labels[labels==-1] = 0
```

```python
plt.scatter(data[labels==-1, 0], data[labels==-1 , 1], c='C0', marker="s", edgecolor='black', linewidth=1)
plt.scatter(data[labels==1, 0], data[labels==1 , 1], c='C1', marker="^", edgecolor='black', linewidth=1)
plt.plot(xlinsp, f(xlinsp), c='C2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Classification Problem')
plt.plot(xlinsp, f(xlinsp), c='C2')
plt.show()
```

```python
estim = LogisticRegression()
estim.fit(data, labels)
```

```python
plot_decision_regions(data, labels, clf=estim, legend=0)
# Adding axes annotations
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Regions')
plt.plot(xlinsp, f(xlinsp), c='C2')
plt.show()
```
