```python
# imports
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

```

```python
n = 100
xmin, xmax = -2, 2
x = np.linspace(xmin, xmax, 100).reshape(n, 1)
# x = np.random.uniform(low=xmin, high=xmax, size=n).reshape(n, 1)
```

```python
f = (lambda x: x**3)

sigma = 1
y = f(x) + np.random.normal(scale=sigma, size=n).reshape(n, 1)
```

```python
plt.scatter(x, y, c='C0', edgecolor='black', linewidth=1)
plt.plot(x, f(x), c='C1')
plt.xticks(())
plt.yticks(())
# plt.xlabel('x')
# plt.ylabel('y')
plt.tight_layout()

plt.savefig('regression_problem.pdf')
plt.show()
```

```python
estim = LinearRegression()
estim.fit(x, y)
```

```python

pred = estim.predict(x)
plt.scatter(x, pred, c='C0', edgecolor='black', linewidth=1)
plt.plot(x, pred, c='C0', linewidth=1)

# Adding axes annotations
plt.xticks(())
plt.yticks(())
# plt.xlabel('x')
# plt.ylabel('y')
plt.plot(x, f(x), c='C1')
plt.tight_layout()

plt.savefig('regression_linearsol.pdf')
plt.show()
```

## Extension of the feature space

```python
x_ext = np.concatenate((x, x**2, x**3), axis=1)
```

```python
estim = LinearRegression()
estim.fit(x_ext, y)
```

```python

pred = estim.predict(x_ext)
plt.scatter(x, pred, c='C0', edgecolor='black', linewidth=1)
plt.plot(x, pred, c='C0', linewidth=1)

# Adding axes annotations
plt.xticks(())
plt.yticks(())
# plt.xlabel('x')
# plt.ylabel('y')
plt.plot(x, f(x), c='C1')
plt.tight_layout()
plt.savefig('regression_cubicsol.pdf')
plt.show()
```
