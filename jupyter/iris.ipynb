```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

import numpy as np

import matplotlib
```

```python
# import some data to play with
iris = load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
# import some data to play with
iris = load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(figsize=(8, 6))
plt.clf()

cmap = matplotlib.cm.get_cmap('Set1')
# Plot the training points
plt.scatter(X[y==0, 0], X[y==0, 1], c=cmap(0), edgecolor="k")
plt.scatter(X[y!=0, 0], X[y!=0, 1], c=cmap(1), edgecolor="k")

plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
# plt.show()
plt.savefig('iris.pdf')
```

```python
plt.figure(2, figsize=(8, 6))
fig, axs = plt.subplots(2, figsize=(8, 6))
# fig.suptitle('Vertically stacked subplots')


axs[0].scatter(X[y==0, 0], np.zeros_like(y[y==0]), c=cmap(0), edgecolor="k")
axs[0].scatter(X[y!=0, 0], np.zeros_like(y[y!=0]), c=cmap(1), edgecolor="k")
axs[0].set_xlabel("Sepal length")
axs[0].set_xticks(())
axs[0].set_yticks(())

axs[1].scatter(X[y==0, 1], np.zeros_like(y[y==0]), c=cmap(0), edgecolor="k")
axs[1].scatter(X[y!=0, 1], np.zeros_like(y[y!=0]), c=cmap(1), edgecolor="k")
axs[1].set_xlabel("Sepal width")
axs[1].set_xticks(())
axs[1].set_yticks(())

# plt.xlim(x_min, x_max)
# plt.show()
plt.savefig('projections_iris.pdf')
```
